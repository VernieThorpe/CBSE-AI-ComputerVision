## **Reflection**

Picture reflection (or mirroring) is a type of scaling that can be used to flip an image vertically or horizontally. We set Sy to -1 and Sx to 1 for reflection along the x-axis, and vice versa for reflection along the y-axis.

The matrix for reflection is given below:

<img src="https://i.ibb.co/PNf3N6z/Unknown-6.png" alt="Unknown-6" border="0">

---

### Python Code:

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# read the input image
img = cv2.imread("ronaldo.jpg")
# convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# disable x & y axis
plt.axis('off')
# show the image
plt.imshow(img)
plt.show()
# get the image shape
rows, cols, dim = img.shape

# transformation matrix for x-axis reflection 
Mx = np.float32([[1,  0, 0   ],
                [0, -1, rows],
                [0,  0, 1   ]])

# transformation matrix for y-axis reflection
My = np.float32([[-1, 0, cols],
                 [ 0, 1, 0   ],
                 [ 0, 0, 1   ]])

# apply a perspective transformation to the image
reflected_imgx = cv2.warpPerspective(img,Mx,(int(cols),int(rows)))
reflected_imgy = cv2.warpPerspective(img,My,(int(cols),int(rows)))

# show the resulting image
plt.imshow(reflected_imgx)
plt.show()
plt.imshow(reflected_imgy)
plt.show()
# save the resulting image to disk
plt.imsave("ronaldo_reflected.jpg", reflected_imgx)
plt.imsave("ronaldo_reflected.jpg", reflected_imgy)
```
